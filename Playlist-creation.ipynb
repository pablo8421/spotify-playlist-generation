```python
#Spotify api connection

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=client_id,
                                               client_secret=client_secret,
                                               redirect_uri=redirect_uri,
                                               scope=scope))
```

```python
# Load pdf with cluster 
songs = pd.read_csv('D:/Pablo/clases/UJM/2. Semester, 2021/Data Mining/Project/tracks_clusterized_metropopolis.csv')

songs.head()
```

```python
# Create playlist
playlist_name = 'Not so happy Metropopolis'

sp.user_playlist_create(username, name=playlist_name)
```

```python
track_ids = songs[songs['cluster'] == 5]['id']

print(track_ids)
```

```python
# Obtained from the response in the creation
playlist_id = 'PLAYLIST-ID-HERE'

# Iterate over a batch size of 50, a maximum of 100 songs can be added for each request
batch_length = 50
for i in range(0,(len(track_ids)+batch_length) // batch_length):
    sp.user_playlist_add_tracks(username, playlist_id, track_ids[batch_length*i:batch_length*i+batch_length])
```
